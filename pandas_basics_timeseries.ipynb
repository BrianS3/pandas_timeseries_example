```python
df['year'] = df['date'].year
```

```python
df['year'] = df['date'].apply(lambda x: x.year)
```

```python
df[['date_time', 'date', 'year']].head(5)
```

```python
df['month'] = df['date'].apply(lambda x: x.month)
```

```python
df['week'] = df['date'].apply(lambda x: x.week)
```

```python
df['week'] = df['date'].apply(lambda x: x.strftime("%V"))
```

```python
df[['date_time', 'date', 'year', 'month', 'week']].head(5)
```

```python
df['time'] = pd.Series([val.time() for val in df['date_time']])
```

```python
print(df['time'][0])
```

```python
str(df['time'][0]) < '22:00:00'
```

```python
df['day/night'] = df['time'].apply( lambda x:  \
                                   'day' if (str(x) > '07:00:00' and str(x) < '19:00:00') else 'night' )
```

```python
df[['date_time', 'date', 'year', 'month', 'week', 'day/night']]
```
